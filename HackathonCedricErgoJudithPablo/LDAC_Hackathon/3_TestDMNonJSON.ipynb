```python
import json
from pprint import pprint

# Read the JSON file and parse it into a list of dictionaries
with open("output/wallsTest.json", "r") as json_file:
    walls_dict = json.load(json_file)

pprint(walls_dict)
```

```python
import requests
import os

endpoint = 'http://127.0.0.1:53081/jitdmn' 
response = requests.post(endpoint, {}, headers={"Content-Type": "application/json"})

# Print the status code
print(f"Status Code: {response.status_code}")
```

```python
# Function to send file and data
def send_file_and_data(i, max_i, dmn, endpoint, data):
    try:
        # Read the XML file
        with open(dmn, 'r', encoding='utf-8') as file:
            xml_data = file.read()

        # Create the request body
        body = {
            "model": xml_data,
            "context": data
        }

        # Send the POST request
        response = requests.post(endpoint, json=body, headers={
            "Content-Type": "application/json"
        })
        
        # Print the response
        print('Response:', response.json())
        response_json = response.json()
        out_file = open("output/testResults.json", "a")
        out_file.write(json.dumps(response_json))
        if not i == max_i:
            out_file.write(',\n')
        out_file.close()
    except Exception as error:
        print('Error sending file and data:', error)

dmn = os.path.join(os.getcwd(), 'data/UValueCheck.dmn') 
out_file = open("output/testResults.json", "w")
out_file.write("[\n")
out_file.close()
i = 0
for s in walls_dict:
    i += 1
    send_file_and_data(i, len(walls_dict), dmn, endpoint, s)

out_file = open("output/testResults.json", "a")
out_file.write("\n]")
out_file.close()
```
