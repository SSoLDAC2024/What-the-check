```python
import json
from rdflib import Graph, Namespace, URIRef

# Load your Turtle file into a graph
g = Graph()
g.parse("output/wallsTest.ttl", format="turtle")
g.print()

# Define namespaces
INST = Namespace("https://linkedbuildingdata.net/ifc/resources20240611_162622/")
PROPS = Namespace("https://w3id.org/props#")
RDF = Namespace("https://www.w3.org/1999/02/22-rdf-syntax-ns#")
RDFS = Namespace("https://www.w3.org/2000/01/rdf-schema#")
EX = Namespace("http://example.org/myData#")
BOT = Namespace("https://w3id.org/bot#")

# Extract the desired information
data = []
#g.subjects(RDF.type, EX.walls)
for sub in g.subjects(RDF.type, EX.walls):
    print(sub)
    name = g.value(sub, RDFS.label)
    thermal_transmittance = g.value(sub, PROPS.Thermaltransmittance)
    guid = g.value(sub, BOT.hasGuid)
    print(name)
    print(thermal_transmittance)
    print(guid)

    if name and thermal_transmittance and guid:
        data.append({
            "name": f"{name}",#:{sub.split('/')[-1]}",
            "Thermaltransmittance": float(thermal_transmittance),
            "guid": str(guid)
        })

# Convert the extracted data to JSON
json_data = json.dumps(data, indent=4)

# Save the JSON data to a file
with open("output/wallsTest.json", "w") as f:
    f.write(json_data)
    f.close()
```
